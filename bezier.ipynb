```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
x = np.array([0.1,1,1.7,2.3])
y = np.array([3.5,1.5,5,2.3])
```

```python
def bezier(v):
    p = []
    angle = lambda p1,p2: math.atan2(p2[1]-p1[1],p2[0]-p1[0])
    for i in range(x.shape[0]):
        p.append([x[i],y[i]])
    while(len(p)-1):
        np = []
        for i in range(len(p)-1):
            dist = math.dist(p[i],p[i+1])
            ang = angle(p[i],p[i+1])
            nx = p[i][0]+math.cos(ang)*dist*v
            ny = p[i][1]+math.sin(ang)*dist*v
            np.append([nx,ny])
        p = np
    return p[0]
```

```python
plt.figure()
t1,t2 = [],[]
for i in range(0,101):
    n = bezier(i/100)
    t1.append(n[0])
    t2.append(n[1])
plt.plot(x,y,label="Without Bezier")
plt.plot(t1,t2,label="Bezier")
plt.legend()
```
